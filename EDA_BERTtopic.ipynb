```python
from bertopic import BERTopic
import pandas as pd
import json
import csv
import plotly.io as pio
```

### Read the csv file

```python
csvFilepath = "Data/TwitterExport/data.csv"
rows = []
with open(csvFilepath, 'r', encoding="utf-8") as csvFile:
    csvReader = csv.reader(csvFile, quoting=csv.QUOTE_NONE, delimiter='\n')
    for row in csvReader:
        if len(row) == 1:
            rows.append(row[0])
print(len(rows))
```

### Generate conversations by usign the line in quotes as the final entry in the conversation

```python
conversations = []
is_test = False
current_conversation = " "
if is_test:
    for row in rows:
        if row[0] == '"':
            current_conversation += row
            conversations.append(current_conversation)
            current_conversation = " "
            if len(conversations) == 10:
                break
        else:
            current_conversation += row
            current_conversation += "\n"

else:
    for row in rows:
        if row[0] == '"':
            current_conversation += row
            conversations.append(current_conversation)
            current_conversation = " "
        else:
            current_conversation += row
            current_conversation += "\n"

print(len(conversations))
```

```python
print(conversations[-1])
conversations = conversations[:1000]
print(len(conversations))
```

```python
topic_model = BERTopic(nr_topics=10, embedding_model="all-MiniLM-L6-v2")
```

```python
topics, probs = topic_model.fit_transform(conversations)
```
